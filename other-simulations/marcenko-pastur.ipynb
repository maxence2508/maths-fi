```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 3000
T = 15000

X = np.random.randn(N, T)
E = (1/T) * X @ X.T
eigenvalues = np.linalg.eigvalsh(E)

q = N / T
lam_m = (1 - np.sqrt(q))**2
lam_p = (1 + np.sqrt(q))**2

def rho_mp(x):
    return np.sqrt((lam_p - x) * (x - lam_m)) / (2 * np.pi * q * x)

# Histogram
plt.figure(figsize=(8,5))
plt.hist(eigenvalues, bins=80, density=True, alpha=0.6, label="Empirical")

# Theoretical curve
x = np.linspace(lam_m, lam_p, 1000)
plt.plot(x, rho_mp(x), linewidth=2, label="Marchenkoâ€“Pastur")

# Vertical dashed lines
plt.axvline(lam_m, linestyle='--', linewidth=2, label=r"$\lambda_-$")
plt.axvline(lam_p, linestyle='--', linewidth=2, label=r"$\lambda_+$")

plt.xlabel("Eigenvalues")
plt.ylabel("Density")
plt.title("Empirical Spectrum vs Marchenko-Pastur Law")
plt.legend()
plt.show()
```
